```python
from keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
% matplotlib inline

(X, Y), (x_test, y_test) = mnist.load_data()
```

```python
print(X.shape)
val_set = np.random.choice(range(len(Y)), 1000, replace=False)
print(val_set.shape)
X_val = X[val_set,:,:]
Y_val = Y[val_set]
X = np.delete(X,val_set,0)
Y = np.delete(Y,val_set,0)
print(X.shape)
print(Y.shape)
```

```python
SZ = 4000
idx_a = []
idx_b = []
idy_a = []
idy_b = []
YYY = []
for i in range(10):
    addr = np.array(np.where(Y==i)).flatten()
    a = np.random.choice(addr, size=SZ, replace=False).tolist()
    b = np.random.choice(addr, size=SZ, replace=False).tolist()
    idx_a = idx_a + a
    idx_b = idx_b + b
    idy_a = idy_a + a
    idy_b = idy_b + b

for i in range(10):
    addr_a = np.array(np.where(Y==i)).flatten()
    addr_b = np.array(np.where(Y!=i)).flatten()
    a = np.random.choice(addr_a, size=SZ, replace=False).tolist()
    b = np.random.choice(addr_b, size=SZ, replace=False).tolist()
    idx_a = idx_a + a
    idx_b = idx_b + b
    idy_a = idy_a + a
    idy_b = idy_b + b
    
XX = [X[idx_a], X[idx_b]]
YY = [Y[idy_a], Y[idy_b]]

for i in range(len(idx_a)):
    YYY.append(0 if YY[0][i]==YY[1][i] else 1)
YYY = np.array(YYY)
```

```python
rng = np.arange(len(idx_a))
train_test = 0.8

# sample ordinal numbers of 
train_id = np.random.choice(rng ,np.round(len(rng)*train_test).astype('int64'), replace=False)
test_id = rng[np.isin(rng,train_id, invert=True)]

x_train = [X[np.array(idx_a)[train_id]], X[np.array(idx_b)[train_id]]]
x_test = [X[np.array(idx_a)[test_id]], X[np.array(idx_b)[test_id]]]
y_train = YYY[train_id]
y_test = YYY[test_id]

print(len(idx_a))
print(len(y_train))
print(len(y_test))
```

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten, Input
from keras.optimizers import Adam
from keras.layers.normalization import BatchNormalization
from keras.utils import np_utils
from keras.layers import Conv2D, MaxPooling2D, ZeroPadding2D, GlobalAveragePooling2D
from keras.layers.advanced_activations import LeakyReLU 
from keras.preprocessing.image import ImageDataGenerator


from keras.layers import Input, Lambda, merge, Dense, DepthwiseConv2D, Activation, AveragePooling2D
from keras.layers import Flatten,Conv2D, MaxPooling2D, Dropout, BatchNormalization, SeparableConv2D
from keras.constraints import max_norm
from keras.models import Model, Sequential
from keras.regularizers import l2
from keras.initializers import RandomUniform
from keras.optimizers import SGD,Adam
from keras.losses import binary_crossentropy
from keras import backend as K

tf.reset_default_graph()
def Mod():
    model = Sequential()
    model.add(Conv2D(32, (3, 3), padding='same', input_shape=(28,28,1)))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(32, (3, 3)))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))

    model.add(Conv2D(64,(3, 3)))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(64, (3, 3)))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))

    model.add(Flatten())

    # Fully connected layer
    model.add(Dense(20))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dropout(0.2))
    model.add(Dense(5))
    return model
```

```python
input_shape = x_train[0][-1,:,:].shape + (1,)
a_input = Input(input_shape)
r_input = Input(input_shape)

encoded_a = Mod()(a_input)
encoded_r = Mod()(r_input)
L1_distance = Lambda(lambda tensors: (K.abs(tensors[0] - tensors[1])))([encoded_a, encoded_r])
L1_distance = Dense(1,activation='sigmoid')(L1_distance)

siamese_net = Model(inputs=[a_input, r_input], outputs=[L1_distance])

optimizer = 'adam'
siamese_net.compile(loss=['binary_crossentropy'], optimizer=optimizer)
siamese_net.summary()
```

```python
train_history = siamese_net.fit([x_train[0][:,:,:,None], x_train[1][:,:,:,None]], y_train,
                epochs=30,
                batch_size=1000,
                verbose=1,
                shuffle=True,
                validation_data=([x_test[0][:,:,:,None], x_test[1][:,:,:,None]], y_test))
```

```python
# print(np.array(history.losses))
import matplotlib.pyplot as plt
%matplotlib inline
loss = train_history.history['loss']
val_loss = train_history.history['val_loss']
plt.plot(loss)
plt.plot(val_loss)
plt.legend(['loss', 'val_loss'])
plt.title('AdaDelta')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.show()
```

```python
rng = np.random.choice(y_train.shape[0],1000, replace=False)
prediction = []
for i in rng:
    pred = siamese_net.predict([x_train[0][None,i,:,:,None], x_train[1][None,i,:,:,None]])
    prediction.append(pred)
    print(i, pred, np.round(pred), y_train[i])
prediction = np.array(prediction).flatten()
prediction = np.round(prediction)
acc = prediction==y_train[rng]
acc = acc*1
print('Train Accuracy', np.mean(acc))
```

```python
rng = np.random.choice(y_test.shape[0],1000, replace=False)
prediction = []
for i in rng:
    pred = siamese_net.predict([x_test[0][None,i,:,:,None], x_test[1][None,i,:,:,None]])
    prediction.append(pred)
    print(i, pred, np.round(pred), y_test[i])
prediction = np.array(prediction).flatten()
prediction = np.round(prediction)
acc = prediction==y_test[rng]
acc = acc*1
print('Test Accuracy', np.mean(acc))
```
